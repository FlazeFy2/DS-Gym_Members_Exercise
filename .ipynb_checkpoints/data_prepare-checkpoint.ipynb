```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("gym_members_exercise_tracking.csv")

# Data Understanding - Variable & Data Type
print(df.dtypes)
print("\n")
print(f"{len(df.columns)} Column")
```

```python
# Data Preparation - Handle Missing Value
df.isnull().sum()
```

```python
# Data Understanding - Descriptive Statistic
df.describe()
```

```python
# Data Preparation - Outliers
ratio_column = ['Age','Weight (kg)','Height (m)','Max_BPM','Avg_BPM','Resting_BPM','Session_Duration (hours)','Calories_Burned','Fat_Percentage','Water_Intake (liters)','Workout_Frequency (days/week)','Experience_Level','BMI']

for dt in ratio_column:
    plt.boxplot(df[dt])
    plt.title(f'Boxplot of {dt}')
    plt.show()
```

```python
# Data Preparation - Invalid Value
ratio_column = ['Age','Weight (kg)','Height (m)','Max_BPM','Avg_BPM','Resting_BPM','Session_Duration (hours)','Calories_Burned','Fat_Percentage','Water_Intake (liters)','Workout_Frequency (days/week)','Experience_Level','BMI']
for dt in ratio_column:
    invalid_prices = df[df[dt] < 0]
    print(invalid_prices.index)
```
