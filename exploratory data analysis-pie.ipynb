```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("gym_members_exercise_tracking.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Most Context for Nonconditional Data
target_col = ['Gender','Workout_Type']

for dt in target_col:
    # Dataset
    top_context = df[dt].value_counts().head(7)
    
    # Pie Chart
    plt.figure(figsize=(8,8))
    top_context.plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f"Most {dt}")
    plt.ylabel("")
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# Exploratory Data Analysis (EDA) - Pie Chart Comparison Total Member by Generation
df = pd.read_csv("gym_members_exercise_tracking.csv")

# Convert age to birth year
current_year = datetime.now().year
df['Birth_Year'] = current_year - df['Age']

# Function to classify generation
def get_generation(year):
    if 1997 <= year <= 2012:
        return "Gen Z"
    elif 1981 <= year <= 1996:
        return "Millennial"
    elif 1965 <= year <= 1980:
        return "Gen X"
    elif 1946 <= year <= 1964:
        return "Boomer"
    else:
        return "Other"

# Create Generation Column
df['Generation'] = df['Birth_Year'].apply(get_generation)

top_generation = df['Generation'].value_counts()

# Pie Chart
plt.figure(figsize=(8,8))
top_generation.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Comparison Total Member by Generation")
plt.ylabel("")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# Dataset
# Exploratory Data Analysis (EDA) - Pie Chart Comparison Total Member by BMI Category - Gender
df = pd.read_csv("gym_members_exercise_tracking.csv")

# Calculate BMI
df['BMI'] = df['Weight (kg)'] / (df['Height (m)'] ** 2)

# Function to classify BMI Category
def get_bmi_category(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 25:
        return "Normal"
    elif 25 <= bmi < 30:
        return "Overweight"
    else:
        return "Obese"

# Create BMI Category Column
df['BMI_Category'] = df['BMI'].apply(get_bmi_category)

# Create Combined Column (BMI Category - Gender)
df['BMI_Gender'] = df['BMI_Category'] + " - " + df['Gender']

# Get Top Context
top_context = df['BMI_Gender'].value_counts().head(7)

# Pie Chart
plt.figure(figsize=(8,8))
top_context.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Comparison Total Member by BMI Category - Gender")
plt.ylabel("")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# Dataset
# Exploratory Data Analysis (EDA) - Pie Chart Comparison Total Water Intake by BMI Category
df = pd.read_csv("gym_members_exercise_tracking.csv")

# Calculate BMI
df['BMI'] = df['Weight (kg)'] / (df['Height (m)'] ** 2)

# Function to classify BMI Category
def get_bmi_category(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 25:
        return "Normal"
    elif 25 <= bmi < 30:
        return "Overweight"
    else:
        return "Obese"

# Create BMI Category Column
df['BMI_Category'] = df['BMI'].apply(get_bmi_category)

# Get Top Context (Total of Water Intake)
top_context = df.groupby('BMI_Category')['Water_Intake (liters)'].sum().sort_values(ascending=False)

# Pie Chart
plt.figure(figsize=(8,8))
top_context.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Comparison Total Water Intake by BMI Category")
plt.ylabel("")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# Dataset
# Exploratory Data Analysis (EDA) - Pie Chart Comparison Total Water Intake by BMI Category - Gender
df = pd.read_csv("gym_members_exercise_tracking.csv")

# Calculate BMI
df['BMI'] = df['Weight (kg)'] / (df['Height (m)'] ** 2)

# Function to classify BMI Category 
def get_bmi_category(bmi):
    if 18.5 <= bmi < 25:
        return "Normal"
    else:
        return "Not Normal"

# Create BMI Category Column
df['BMI_Category'] = df['BMI'].apply(get_bmi_category)

# Combine BMI Category with Gender
df['BMI_Gender'] = df['BMI_Category'] + " - " + df['Gender']

# Get Top Context (SUM of Water Intake)
top_context = df.groupby('BMI_Gender')['Water_Intake (liters)'].sum().sort_values(ascending=False)

# Pie Chart
plt.figure(figsize=(8,8))
top_context.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Comparison Total Water Intake by BMI Category and Gender")
plt.ylabel("")
plt.show()
```
