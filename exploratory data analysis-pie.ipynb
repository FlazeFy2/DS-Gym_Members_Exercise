```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("gym_members_exercise_tracking.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Most Context for Nonconditional Data
target_col = ['Gender','Workout_Type']

for dt in target_col:
    # Dataset
    top_context = df[dt].value_counts().head(7)
    
    # Pie Chart
    plt.figure(figsize=(8,8))
    top_context.plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f"Most {dt}")
    plt.ylabel("")
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# Exploratory Data Analysis (EDA) - Pie Chart Comparison Total Member by Generation - Gender
df = pd.read_csv("gym_members_exercise_tracking.csv")

# Convert age to birth year
current_year = datetime.now().year
df['Birth_Year'] = current_year - df['Age']

# Function to classify generation
def get_generation(year):
    if 1997 <= year <= 2012:
        return "Gen Z"
    elif 1981 <= year <= 1996:
        return "Millennial"
    elif 1965 <= year <= 1980:
        return "Gen X"
    elif 1946 <= year <= 1964:
        return "Boomer"
    else:
        return "Other"

# Create Generation Column
df['Generation'] = df['Birth_Year'].apply(get_generation)

# Create Combined Column (Generation - Gender)
df['Gen_Gender'] = df['Generation'] + " - " + df['Gender']

top_context = df['Gen_Gender'].value_counts().head(7)

# Pie Chart
plt.figure(figsize=(8,8))
top_context.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Comparison Total Member by Generation - Gender")
plt.ylabel("")
plt.show()
```
